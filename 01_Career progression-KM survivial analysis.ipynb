```python
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter

# Load data
DATA_DIR = Path("../data")
df = pd.read_excel( DATA_DIR/demo_career_progression,
                   sheet_name='PSE_R3_new')

T = df["time_R3"]       # survival time
E = df["Status_R3"]     # Event indicator（1=process to R3，0= no process）
groups = df["Type"]     # Type：Abroad / Visiting

ix_abroad = (groups == "Abroad PhDs")
ix_visit  = (groups == "Visiting PhDs")

kmf = KaplanMeierFitter()

# ===== Layout setting =====
plt.rcParams.update({
    "font.family": "Times New Roman",
    "font.size": 18,
    "axes.labelsize": 18,
    "axes.titlesize": 18,
    "legend.fontsize": 16,
    "xtick.labelsize": 18,
    "ytick.labelsize": 18,
})

fig, ax = plt.subplots(figsize=(8, 6))

def plot_cum(T_sub, E_sub, label, color, t_max=15):
    
    kmf.fit(T_sub, event_observed=E_sub, label=label)

    # KM survival analysis 
    sf = kmf.survival_function_.iloc[:, 0]          # Series: S(t)
    ci = kmf.confidence_interval_survival_function_
    S_lower = ci.iloc[:, 0]                         
    S_upper = ci.iloc[:, 1]                         

    # F(t) = 1 - S(t)
    F = 1 - sf                                     
    F_lower = 1 - S_upper                           
    F_upper = 1 - S_lower                           

   
    last_t = F.index[-1]
    last_val = F.iloc[-1]
    last_low = F_lower.iloc[-1]
    last_up  = F_upper.iloc[-1]

    if last_t < t_max:
        F.loc[t_max] = last_val
        F_lower.loc[t_max] = last_low
        F_upper.loc[t_max] = last_up

        # sort by time
        F = F.sort_index()
        F_lower = F_lower.sort_index()
        F_upper = F_upper.sort_index()
    # =========================

    # main curve
    ax.step(
        F.index, F.values,
        where="post",
        color=color,
        linewidth=2,
        label=label,
        zorder=3
    )

    
    ax.fill_between(
        F.index,
        F_lower.values,
        F_upper.values,
        color=color,
        alpha=0.18,
        linewidth=0,
        zorder=2
    )


plot_cum(T[ix_abroad], E[ix_abroad],
         label="Abroad PhDs in PSE",
         color="#EE6363",   # 红
         t_max=15)

plot_cum(T[ix_visit], E[ix_visit],
         label="Visiting PhDs in PSE",
         color="#228B22",   # 绿
         t_max=15)


ax.set_xlabel("Years since PhD graduation")
ax.set_ylabel("Proportion of researchers promoted to R3")

ax.set_ylim(0, 1.0)
ax.set_xlim(0, 15)
ax.set_xticks(range(0, 16, 1))

ax.grid(axis="y", alpha=0.25, linestyle="--", linewidth=0.7)
ax.grid(axis="x", alpha=0.25, linestyle="--", linewidth=0.7)

ax.legend(frameon=False, loc="upper left")

fig.tight_layout()

plt.savefig("E:/P4/KM_cumulative_R3_PSE.png",
            dpi=300, bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter


# Load data
DATA_DIR = Path("../data")
df = pd.read_excel( DATA_DIR/demo_career_progression,
                   sheet_name='PSE_R4_new')


T = df["time_R4"]       
E = df["Status_R4"]      
groups = df["Type"]      


ix_abroad = (groups == "Abroad PhD")
ix_visit  = (groups == "Visiting PhD")

kmf = KaplanMeierFitter()

plt.rcParams.update({
    "font.family": "Times New Roman",
    "font.size": 18,
    "axes.labelsize": 18,
    "axes.titlesize": 18,
    "legend.fontsize": 16,
    "xtick.labelsize": 18,
    "ytick.labelsize": 18,
})

fig, ax = plt.subplots(figsize=(8, 6))

def plot_cum(T_sub, E_sub, label, color):
    kmf.fit(T_sub, event_observed=E_sub, label=label)
    sf = kmf.survival_function_
    ci = kmf.confidence_interval_survival_function_

    F = 1 - sf
    F_lower = 1 - ci.iloc[:, 1]
    F_upper = 1 - ci.iloc[:, 0]

    
    ax.step(
        F.index, F.values.flatten(),
        where="post",
        color=color,
        linewidth=2,
        label=label,
        zorder=3
    )


    ax.fill_between(
        F.index,
        F_lower.values.flatten(),
        F_upper.values.flatten(),
        color=color,
        alpha=0.18,
        linewidth=0,
        zorder=2
    )


plot_cum(T[ix_abroad], E[ix_abroad], "Abroad PhDs in PSE", "#EE6363")
plot_cum(T[ix_visit],  E[ix_visit],  "Visiting PhDs in PSE", "#228B22")


ax.set_xlabel("Years since PhD graduation")
ax.set_ylabel("Proportion of researchers promoted to R4")


ax.set_ylim(0, 1.0)
ax.set_xlim(0, 15)    
ax.set_xticks(range(0, 16, 1))




ax.grid(axis="y", alpha=0.25, linestyle="--", linewidth=0.7)
ax.grid(axis="x", alpha=0.25, linestyle="--", linewidth=0.7)

###
ax.legend(frameon=False, loc="upper left")

fig.tight_layout()

plt.savefig("E:/P4/KM_cumulative_R4_PSE1.png", dpi=300, bbox_inches="tight")
plt.show()
```
