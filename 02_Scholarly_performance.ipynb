```python
###PSE number (fractional counting)
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio
 
# Load data
DATA_DIR = Path("../data")
df = pd.read_excel(DATA_DIR /'demo_scholarly_performance.xlsx', sheet_name='Fractional counting PSE new')
 
# Define color scheme
color_scheme = {
    'Abroad PhDs in PSE': {
        'line': '#EE6363',
        'fill': 'rgba(238, 99, 99, 0.2)'
    },
    'Visiting PhDs in PSE': {
        'line': '#228B22',
        'fill': 'rgba(34, 139, 34, 0.2)'
    }
}
 
# Initialize figure
fig = go.Figure()
 
# Plot for each group
for t in df['Type'].unique():
    sub_df = df[df['Type'] == t]
    x_vals = sub_df['year_gap']
    y_vals = sub_df['agvpapersperyear']
    se_vals = sub_df['se']
 
    # Error band
    fig.add_trace(go.Scatter(
        x=list(x_vals) + list(x_vals[::-1]),
        y=list(y_vals - se_vals) + list(y_vals + se_vals)[::-1],
        fill='toself',
        fillcolor=color_scheme[t]['fill'],
        line=dict(color='rgba(255,255,255,0)'),
        hoverinfo="skip",
        showlegend=False
    ))
 
    # Main line
    fig.add_trace(go.Scatter(
        x=x_vals,
        y=y_vals,
        mode='lines',
        name=t,
        line=dict(color=color_scheme[t]['line'], width=2),
        line_shape='spline',
        hovertemplate='Year Gap: %{x}<br>MNCS: %{y:.2f}<extra>' + t + '</extra>'
    ))
 
# Update layout
fig.update_layout(
    title=dict(
        text='<b>Pubs Over Time with Error Bands</b>',  
        font=dict(family='Times New Roman', size=18,color='black'),  
        x=0.5,
        xanchor='center'
    ),
    xaxis_title=dict(
        text='<b>Year before/after PhD graduation</b>',
        font=dict(family='Times New Roman', size=18, color='black')
    ),
    yaxis_title=dict(
        text='<b>APNR</b>',
        font=dict(family='Times New Roman', size=18, color='black')
    ),
    
    legend=dict(
        font=dict(family='Times New Roman', size=18,color='black')
    ),
    template='plotly_white',
    margin=dict(l=60, r=40, t=60, b=40),
    font=dict(family='Times New Roman', size=18),  
    shapes=[
        dict(
            type='line',
            x0=0, x1=0,
            yref='paper', y0=0, y1=1,
            line=dict(color='black', width=1.5, dash='dot'),
            layer='below'
        ),
        dict(
            type='line',
            x0=6.3, x1=6.3,
            yref='paper', y0=0, y1=1,
            line=dict(color='gray', width=1.5, dash='dot'),
            layer='below'
        )
    ],
    xaxis=dict(
        tickmode='linear',
        dtick=1,
        zeroline=False,
        ticks='outside',
        mirror=True,
        linewidth=1.2,
        linecolor='black',
        tickfont=dict(family='Times New Roman', size=18,color='black'),
    ),
    yaxis=dict(
        range = [0,1.5],
        ticks='outside',
        mirror=True,
        linewidth=1.2,
        linecolor='black',
        tickfont=dict(family='Times New Roman', size=18,color='black'),
    )
)
 
# Show figure
fig.show()
 
# Save as high-resolution PNG
fig.write_image("../pubs_over_time_PSE_new.png", width=700, height=400, scale=3)
```
