```python
####PSE no collaboration

import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio

# Load data
DATA_DIR = Path("../data")
df = pd.read_excel(DATA_DIR / "demo_academic_embeddedness.xlsx", sheet_name= 'PSE 0')

# Define color scheme
color_scheme = {
    'Abroad PhDs in PSE': {
        'line': '#EE6363',      
        'fill': 'rgba(238, 99, 99, 0.2)'     
    },
    'Visiting PhDs in PSE': { 
        'line': '#228B22',      
        'fill': 'rgba(34, 139, 34, 0.2)'     
    }
}

# Initialize figure
fig = go.Figure()

# Plot for each group
for t in df['Type'].unique():
    sub_df = df[df['Type'] == t]
    x_vals = sub_df['gap_year']
    y_vals = sub_df['avg_proportion']
    se_vals = sub_df['standard_error']

    # Error band
    fig.add_trace(go.Scatter(
        x=list(x_vals) + list(x_vals[::-1]),
        y=list(y_vals - se_vals) + list(y_vals + se_vals)[::-1],
        fill='toself',
        fillcolor=color_scheme[t]['fill'],
        line=dict(color='rgba(255,255,255,0)'),  
        hoverinfo="skip",
        showlegend=False
    ))

    # Main line
    fig.add_trace(go.Scatter(
        x=x_vals,
        y=y_vals,
        mode='lines',
        name=f'{t}',
        line=dict(color=color_scheme[t]['line'], width=2),
        line_shape='spline',
        hovertemplate='Year Gap: %{x}<br>MNCS: %{y:.2f}<extra>' + t + '</extra>'
    ))

# Layout settings
    
    fig.update_layout( 
        xaxis_title=dict(
        text='<b>Year before/after PhD graduation</b>',
        font=dict(family='Times New Roman', size=14,color='black')
    ),
    
    yaxis_title=dict(
        text='<b>PP (no collab)</b>',
        font=dict(family='Times New Roman', size=14, color='black')
    ),
    yaxis_tickformat='.0%',
   
    legend=dict(
        font=dict(family='Times New Roman', size=14,color='black')
    ),
    template='plotly_white',
    font=dict(family="Times New Roman", size=14, color='black' ),
    margin=dict(l=60, r=40, t=60, b=40),

    
    shapes=[
        
        dict(
            type='line',
            x0=0.01, x1=0.01,
            yref='paper', y0=0, y1=1,
            line=dict(color='black', width=1.5, dash='dot'),
            layer='below'
        )
    ],

    
    xaxis=dict(
        tickmode='linear',
        dtick=1,
        #showgrid=True,
        zeroline=False,
        ticks='outside',
        #showline=True,
        mirror=True,
        linewidth=1.2,
        linecolor='black',
        tickfont=dict(family="Times New Roman",size=14, color='black')
    ),
    yaxis=dict(
        #showgrid=True,
        ticks='outside',
        #showline=True,
        mirror=True,
        linewidth=1.2,
        linecolor='black',
        tickfont=dict(family="Times New Roman",size=14,color='black')
    )

)
# Show plot
fig.show()
# Save as high-resolution PNG
fig.write_image("../NO_Collaboration_PSE.png", width=600, height=400, scale=3)
```
